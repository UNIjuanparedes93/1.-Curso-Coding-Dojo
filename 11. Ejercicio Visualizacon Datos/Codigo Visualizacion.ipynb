```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = r'C:/Users/Juan.Paredes/Unilever/Data & Analytics CL DEV - General/Nacho/1. Visual Studio Scripts/CURSO DOJO/1.-Curso-Coding-Dojo/11. Ejercicio Visualizacon Datos/USA Popular Sports.xlsx'
df = pd.read_excel(io= file, engine= None)
df
```

```python
# Hacemos una pequeña transformacion de las columnas para pasarlas de columnas a filas.

df = pd.melt(df, id_vars='Sport', value_vars=[2017, 2013, 2008, 2007, 2006, 2005, 2004], var_name='Year', value_name='% Total')
df
```

```python
# Definimos variables
labels = df['Year'].unique()
football_total = df['% Total'][df['Sport'] == 'Football'].to_list()
basketball_total = df['% Total'][df['Sport'] == 'Basketball'].to_list()
baseball_total = df['% Total'][df['Sport'] == 'Baseball'].to_list()


width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

# Variables a aparecer en las barras
p1 = ax.bar(labels, football_total, width, label='Football')
p2 = ax.bar(labels, basketball_total, width,label='Basketball')
p3 = ax.bar(labels, baseball_total, width,label='Baseball',bottom=basketball_total)

# Ponemos los labels
ax.bar_label(p1, label_type='edge')
ax.bar_label(p2, label_type='edge')
ax.bar_label(p3,label_type='edge')

# Ejes
ax.set_ylabel('% Popularity')
ax.set_title('US Top 3 most popular sports by Year')
ax.legend()

plt.show()
```
